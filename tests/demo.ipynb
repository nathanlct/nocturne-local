```python
import numpy as np
import matplotlib.pyplot as plt

from nocturne import Simulation
```

```python
sim = Simulation(scenarioPath='../scenarios/intersection.json')
scenario = sim.getScenario()
```

```python
objects = scenario.getRoadObjects()
print(objects)

veh = objects[-1]
```

```python
coneImages = []
goalImages = []
n = 20

for i in range(n):
    coneImg = np.array(scenario.getCone(veh, 3.14/2.0, 0.0), copy=False)
    coneImages.append(coneImg)
    
    goalImg = np.array(scenario.getGoalImage(veh), copy=False)
    goalImages.append(goalImg)
    
    veh.setAccel(20.0 if i <= 8 else 0.0)
    veh.setSteeringAngle(-12. * 3.1415 / 180.0 if i >= 6 and i <= 9 else 0)
    
    for _ in range(5): #Â 0.5s
        sim.step(0.1)      
```

```python
plt.figure()


fig, axes = plt.subplots(n, 2, figsize=(15, n * 7.5)) 

for i in range(n):
    axes[i][0].imshow(coneImages[i])
    axes[i][1].imshow(goalImages[i])
```
